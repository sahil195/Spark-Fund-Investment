```python
# Importing data from rounds2.csv file using encoding 'ISO-8859-1'

rounds2 = pd.read_csv("rounds2.csv", encoding = "ISO-8859-1")
rounds2
```

```python
# Converting company_permalink column of rounds2 dataframe to lower case in order to 
# 1) find unique companies
# 2) Merging happens correctly with the companies dataframe

#Why are we doing this?
#The file which we're able to download from user Aditya Chopra gives us company_permalink as both lowercase and uppercase values.
#The file which we're able to download from user Sahil Bansal gives us company_permalink column as lowercase values only.

#Thus, in order to avoid incorrect results while evaluation we're converting it to lower case


# Moreover, If we don't convert it to lowercase the unique() function counts "abc" and "ABC" as two different values
# Either we use ignoreCase() or we convert it to lowercase()
# Proceeding with the lowercase approach

rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
rounds2
```

```python
# deleting nan values rows in category_list
master_frame=master_frame.loc[~pd.isnull(master_frame['category_list'])]
master_frame
```

```python
#separating primary sectors from the category list in master_frame
def splitLogic(str):
    splittedCategories = str.split('|')
    if(len(splittedCategories)>=1):
        return( splittedCategories[0])

    
master_frame['primary_sector'] = master_frame['category_list'].apply(splitLogic)
master_frame[['category_list','primary_sector']]
```

```python
#importing data from mapping.csv file
mapping = pd.read_csv("mapping.csv")
mapping
```

```python
#dropping blank column as blank category is not required
mapping=mapping.drop('Blanks',axis=1)
mapping
```

```python
#checking nan values
mapping.isnull().sum()
```

```python
#removing rows with NaN value in 'category_list' column
mapping=mapping.loc[~pd.isnull(mapping['category_list'])]
mapping
```

```python
#importing mapping.csv file
mapping = pd.read_csv("mapping.csv")

# replacing '0' at the start position with 'Na' in the mapping dataframe in order to correctly find out the main sector corresponding to each primary sector
mappingReplaceFront = mapping.replace(to_replace='^0', value = 'Na', regex=True)

#Checking that replacement has been done correctly or not
mappingReplaceFront.loc[mappingReplaceFront['category_list'] == 'Nanotechnology']

#Replacing '0' not at first position with 'na' in the mapping dataframe in order to correctly find out the main sector corresponding to each primary sector
mappingReplaceBetween = mappingReplaceFront.replace(to_replace = '0', value = 'na', regex=True)
mappingReplaceBetween
```

```python
#Rectifying the affected values as below so that the later on correct main_sector could be found corresponding to primary_sector
mappingReplaceBetween = mappingReplaceBetween.replace(to_replace='Enterprise 2.na', value='Enterprise 2.0', regex=True)
mappingReplaceBetween
```

```python
# melting the mapping dataframe
melted = pd.melt(mappingReplaceBetween, id_vars=["category_list"], var_name="main_sector", value_name="score")
melted
```

```python
#Dropping rows with 0 score and thus getting exact mapping of category_list to primary_sector
melted = melted.loc[melted['score'] == 1]
melted
```

```python
#dropping score column
melted=melted.drop('score',axis=1)
melted
```

```python
#Checking melted DF for null values
melted.isnull().sum()
```

```python
#Checking for proper replacement of '0' in the melted Dataframe
melted.loc[melted['category_list'] == 'Enterprise 2.0']
```

```python
# merged dataframe
merged = pd.merge(master_frame, melted, how='left', left_on = 'primary_sector',right_on ='category_list')
merged
```

```python
#removing rows containing NaN values in main_sector column
mergednulldeleted = merged.loc[~pd.isnull(merged['main_sector'])]
mergednulldeleted
```

```python
#creating D1
D1 = merged.loc[(merged['country_code'] == 'USA') & (merged['funding_round_type'] == 'venture') ]
D1=D1.loc[(D1['raised_amount_usd'] >=   5000000) & (D1['raised_amount_usd'] <= 15000000) ]
D1['total_invested_amount'] = D1.groupby(['main_sector'])['raised_amount_usd'].transform('sum')
D1['total_investment_number'] = D1.groupby(['main_sector'])['raised_amount_usd'].transform('count')
D1
# D1.to_csv(r"C:\Users\Lenovo\Desktop\D1.csv", encoding='utf-8')
```

```python
# creating D2
D2=merged.loc[(merged['country_code'] == 'GBR') & (merged['funding_round_type'] == 'venture') ]
D2=D2.loc[(D2['raised_amount_usd'] >= 5000000) & (D2['raised_amount_usd'] <= 15000000) ]
D2['total_invested_amount'] = D2.groupby(['main_sector'])['raised_amount_usd'].transform('sum')
D2['total_investment_number'] = D2.groupby(['main_sector'])['raised_amount_usd'].transform('count')
D2
# D2.to_csv(r"C:\Users\Lenovo\Desktop\D2.csv", encoding='utf-8')
```

```python
#CREATING D3
D3=merged.loc[(merged['country_code'] == 'IND') & (merged['funding_round_type'] == 'venture') ]
D3=D3.loc[(D3['raised_amount_usd'] >= 5000000) & (D3['raised_amount_usd'] <= 15000000) ]
D3['total_invested_amount'] = D3.groupby(['main_sector'])['raised_amount_usd'].transform('sum')
D3['total_investment_number'] = D3.groupby(['main_sector'])['raised_amount_usd'].transform('count')
D3
# D3.to_csv(r"C:\Users\Lenovo\Desktop\D3.csv", encoding='utf-8')
```

```python
#concating D1,D2 AND D3 and exporting to CSV file
combined=pd.concat([D1,D2,D3])
combined
combined.to_csv(r"C:\Users\Aditya Chopra\Desktop\combined.csv", encoding='utf-8')
```
